# Data integration

```python
import pandas as pd
# reloding the dataset to check if the changes are permanent
file_naam = "https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/amazon_fires.csv"
df = pd.read_csv(file_naam, encoding='latin-1')    
new_columns_names = {
    'ano' : 'year',
    'mes' : 'month', 
    'estado' : 'state',
    'numero': 'number_of_fires',    
    'encontro': 'date'
}
df.rename(columns = new_columns_names, inplace = True)
df['state'] = df['state'].str.title()

df.head()
df['number_of_fires'] = df['number_of_fires'].str.strip("Fires")
df['number_of_fires'] = df['number_of_fires'].fillna(0)
df['number_of_fires'] = df['number_of_fires'].astype(float)

# introducing the map function
month_translations = {
    'Janeiro': 'January',
    'Fevereiro': 'February',
    'MarÃ§o': 'March',

    'Abril': 'April',
    'Maio': 'May',
    'Junho': 'June',
    'Julho': 'July',
    'Agosto': 'August',
    'Setembro': 'September',
    'Outubro': 'October',
    'Novembro': 'November',
    'Dezembro': 'December'
}
df['month'] = df['month'].map(month_translations)
df.head(20)
```
