```python
# Loading the dataset 
import pandas as pd

file_name = "https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/amazon_fires.csv"
df = pd.read_csv(file_name, encoding='latin-1')
df.head()
```

```python
#To check the regions in the dataset
df['estado'].unique()
```

# Renaming the columns to English

```python
new_columns_names = {
    'ano' : 'year',
    'mes' : 'month',
    'estado' : 'state',
    'numero': 'number_of_fires',
    'encontro': 'date'
}
df.rename(columns = new_columns_names, inplace = True)
```

```python
df.head()
```

```python
df['year'].unique()
```

```python
df.info()
```

# Rearranging the columns

```python
# reordering the columns accordingly
new_order = [4,1,0,2,3]
df = df[df.columns[new_order]]
df.head()
```

# Determining if a column contains numeric data

```python
# it isn't, let's find our why
df['number_of_fires'].str.isnumeric()
```

```python
df['number_of_fires'].astype(str).str.isdigit()
```

# Removing the unecessay text from the columns

```python
# using strip to remove the spaces and then checking if the values are numeric
df['number_of_fires'].str.strip("Fires")
```

```python
# Replacing the column with the cleaned column
df['number_of_fires'] = df['number_of_fires'].str.strip("Fires").astype(int)
df.head()
```

```python
df.info()
```

```python
# conveting the number of fires to a float data type and also this is the alternative string manipulation method to remove the "Fires" string from the column
# df["number_of_fires"] = df["number_of_fires"].str.replace("", "0").astype(float)
# df.head()
print(df["number_of_fires"].dtype)
print(df["number_of_fires"].head())
```

```python
# If it's already numeric â just convert directly, no str needed
df["number_of_fires"] = df["number_of_fires"].astype(float)

# If it's mixed (some strings, some numbers) â convert safely
df["number_of_fires"] = pd.to_numeric(df["number_of_fires"], errors='coerce').fillna(0)
```

```python
df.head()
```

# Handelling the missing values

```python
# reloding the dataset
file_naam = "https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/amazon_fires.csv"
df = pd.read_csv(file_naam, encoding='latin-1')
naya_columns = {
    'ano' : 'year',
    'mes' : 'month',
    'estado' : 'state',
    'numero': 'number_of_fires',
    'encontro': 'date'
}
df.rename(columns = naya_columns, inplace = True)
df['number_of_fires'] = df['number_of_fires'].str.strip("Fires")
#creating a true copy of the dataset
df_copy = df.copy()
df_copy.head()
```

```python
df_copy.info()
```

```python
df_copy.isnull().sum()
```

```python
#  we easily remove the NUll or Nan (not a number) values from the dataset using the dropna() method    
df_copy.dropna(inplace = True)
df_copy.reset_index()
df_copy.head()
```

```python
df_copy.isnull().sum()
```

# Other MEthods of handelling missing values

```python
# reloding the dataset to check if the changes are permanent
file_naam = "https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/amazon_fires.csv"
df = pd.read_csv(file_naam, encoding='latin-1')    
new_columns_names = {
    'ano' : 'year',
    'mes' : 'month', 
    'estado' : 'state',
    'numero': 'number_of_fires',    
    'encontro': 'date'
}
df.rename(columns = new_columns_names, inplace = True)

df['number_of_fires'] = df['number_of_fires'].str.strip("Fires")
df_kapy = df.copy()
df.head()
```

```python
df.isnull().sum()
```

```python
df[df['number_of_fires'].isnull()].head ()
```

##### what to do with the missing data? 
1. remove them using dropn(axis = 0 )
2 . replace them with some arbitary number (e.g. an average)
3 . replace them zeros , or forward fill method or backward fill 

```python
#using fillna with zeros
df = df_kapy
df['number_of_fires'] = df['number_of_fires'].fillna(0)
df.head(70)
```

```python
# lets try with the filling 
df['number_of_fires'] = df['number_of_fires'].fillna(method = 'ffill')
df.head(70)
```
